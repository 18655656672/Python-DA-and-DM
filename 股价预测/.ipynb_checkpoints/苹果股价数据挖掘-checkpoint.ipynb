## 导入数据

```python
import pandas as pd 
from datetime import datetime

"""数据导入

这里我们将会导入2015年8月3日苹果公司的秒级股票交易数据，这里的原始数据
需要进行稍许清理才能使用，正好符合我们本章节的学习要点。
"""

## 这里我们用了Pandas的read_csv模块直接从csv文件中导入数据。

data = pd.read_csv("aapl.csv",
                   names = ["timestamp_raw","Open","High",
                            "Low","Close","Volume"],
                   index_col = False)
data.head(5)
```

## 数据清洗

```python
## 原始数据中的时间记录为每天距离格林威治标准时间的秒数乘以1000，为增加
## 可读性，我们需要将数据还原。这里我们让data对象的索引变为了处理后的时
## 间标记，调用了DataFrame.index域。

UNIX_EPOCH = datetime(1970, 1, 1, 0, 0)
def ConvertTime(timestamp_raw, date):
    """ 该函数会将原始的时间转化为所需的datetime格式 """
    delta = datetime.utcfromtimestamp(timestamp_raw) - UNIX_EPOCH
    return date + delta   

data.index = map(lambda x: ConvertTime(x, datetime(2015, 8, 3)),
                 data["timestamp_raw"]/1000)

## 这个时候timestamp_raw一列不在有用，我们可以删掉它。
data = data.drop("timestamp_raw",1)

## 我们可以看到交易时间其实包括了盘前和盘后大量时间。在实际交易策略中，
## 我们往往只会在正常交易时间进行交易。所以需要对数据按照时间进行拆分。
## 这样的操作非常容易：

data_trading_hour = data["201508030930":"201508031529"]
## 将交易时间的数据保存起来供后面使用：
data_trading_hour.to_csv("aapl-trading-hour.csv")#将处理后的数据保存，用以后续建模调用
```

## 可视化原始数据

```python
import matplotlib
from matplotlib import pyplot as plt
plt.rcParams['font.sans-serif']='SimHei'#设置中文字符的正常显示
plt.rcParams['axes.unicode_minus'] = False
## 设置画图风格为R的ggplot风
matplotlib.style.use('ggplot')

## 每秒收盘价是什么样子的走势？可以这么看

data_trading_hour["Close"].plot()
plt.title('股价走势（每秒收盘价）')
## plt.savefig("plots/closing-price-plot.png")
plt.show()

"""收盘价变化率

当然做实时量化交易我们最感兴趣的还是每秒的变化率。那么我们来看看股价变
化有什么样的分布。
"""

change = data_trading_hour["Close"].diff()/data_trading_hour["Close"]
change.plot()
plt.title('秒级收盘价变化率')
## plt.savefig("plots/price-change-plot.png")
plt.show()

## 看看前一秒变化率和后一秒变化率之间的关系。
print(change.shift(1).corr(change))
print(change.shift(2).corr(change))
```

## 建立预测模型

```python
import pandas as pd 
import numpy as np
from sklearn.preprocessing import Imputer, PolynomialFeatures
from sklearn.pipeline import Pipeline, FeatureUnion
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score, median_absolute_error,mean_squared_error 
from timeseriesutil import TimeSeriesDiff, TimeSeriesEmbedder, ColumnExtractor
from sklearn.neural_network import  MLPRegressor 
from sklearn.svm import SVR
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt 
import matplotlib 
import warnings 

warnings.filterwarnings("ignore")#忽视警告
matplotlib.style.use('ggplot')#设置绘图风格
plt.xticks(rotation=90)#设置x轴标签旋转90度

"""数据导入

这里我们导入经过了预处理的苹果公司股票报价，该数据仅包含正常交易时间的
数据。
"""

data = pd.read_csv("aapl-trading-hour.csv",index_col = 0)

y = data["Close"].diff() / data["Close"].shift()#股价变化率
y[np.isnan(y)]=0#将缺失值赋值为0

n_total = data.shape[0]#样本个数
n_train = int(np.ceil(n_total*0.7))#划分70%的数据作为训练集

data_train = data[:n_train]
data_test  = data[n_train:]

y_train = y[10:n_train]
y_test  = y[(n_train+10):]

#选取close和volume作为特征变量
'''
管道进行预处理的四个步骤
1.选取特征列
2.作时间序列差分
3.将2中的结果转换为列表形式
4.填补上述计算中产生的缺失值
'''
#将price列 进行数据预处理
pipeline_closing_price= Pipeline([("ColumnEx", ColumnExtractor("Close")),
                                   ("Diff", TimeSeriesDiff()),
                                   ("Embed", TimeSeriesEmbedder(10)),
                                   ("ImputerNA", Imputer())])
#将volume列进行数据预处理
pipeline_volume = Pipeline([("ColumnEx", ColumnExtractor("Volume")),
                            ("Diff", TimeSeriesDiff()),
                            ("Embed", TimeSeriesEmbedder(10)),
                            ("ImputerNA", Imputer())])


#合并特征
merged_features = FeatureUnion([("ClosingPriceFeature", pipeline_closing_price),
                                ("VolumeFeature", pipeline_volume)])
#合并管道并建立模型
pipeline_2 = Pipeline([("MergedFeatures", merged_features),
                       ("PolyFeature",PolynomialFeatures()),
                       ('linGer',LinearRegression())])
pipeline_2.fit(data_train, y_train)
y_pred_2 = pipeline_2.predict(data_test)

#可视化结果
plot_train=(y_train+1)* data["Close"].shift()
plot_train.plot(style="b", rot=10)#绘制测试集的数据

plot_test=(y_test+1)* data["Close"].shift()
plot_test.plot(style="y", rot=10)#绘制测试集的数据

cc_2 = np.sign(y_pred_2)*y_test#每笔交易的理论收益
cumulative_return_2 = (cc_2+1).cumprod()* data["Close"].shift()
cumulative_return_2.plot(style="r", rot=10)#绘制预测的数据
plt.legend(['训练集','测试集','预测数据'])
plt.title('股价预测模型')

#模型评估
print("R²:",r2_score(y_test, y_pred_2))
print("平均绝对值误差",median_absolute_error(y_test, y_pred_2))
print("均方误差",mean_squared_error(y_test,y_pred_2))
```

         由上述结果可以看到，使用线性回归对结果的预测非常不理想，偏差非常大，主要原因在于股价的走势会受很多市场因素左右。我们接下来尝试使用不同的模型来拟合数据来观察
     是否会有同样大的偏差。

## 封装上述的建模预测过程

```python
#选择特征变量收盘价和成交量
def build_model(name,model):
    pipeline_closing_price = Pipeline([("ColumnEx", ColumnExtractor("Close")),
                                       ("Diff", TimeSeriesDiff()),
                                       ("Embed", TimeSeriesEmbedder(10)),
                                       ("ImputerNA", Imputer())])

    pipeline_volume = Pipeline([("ColumnEx", ColumnExtractor("Volume")),
                                ("Diff", TimeSeriesDiff()),
                                ("Embed", TimeSeriesEmbedder(10)),
                                ("ImputerNA", Imputer())])

    #合并上述两个特征变量，作为模型的输入
    merged_features = FeatureUnion([("ClosingPriceFeature", pipeline_closing_price),
                                    ("VolumeFeature", pipeline_volume)])
    
    #建立模型
    pipeline_2 = Pipeline([("MergedFeatures", merged_features),
                           ("PolyFeature",PolynomialFeatures()),
                            ('StandardScaler',StandardScaler()),
                           ("LinReg",model)])

    pipeline_2.fit(data_train, y_train)
    y_pred_2 = pipeline_2.predict(data_test)#预测结果


    #可视化结果
    plot_train=(y_train+1)* data["Close"].shift()
    plot_train.plot(style="b", rot=10)#绘制测试集的数据

    plot_test=(y_test+1)* data["Close"].shift()
    plot_test.plot(style="y", rot=10)#绘制测试集的数据

    cc_2 = np.sign(y_pred_2)*y_test#每笔交易的理论收益
    cumulative_return_2 = (cc_2+1).cumprod()* data["Close"].shift()
    cumulative_return_2.plot(style="r", rot=10)#绘制预测的数据
    plt.legend(['训练集','测试集','预测数据'])
    plt.title('股价变化预测模型--'+name)
    plt.show()
    
    
    #模型评估
    print("R²:",r2_score(y_test, y_pred_2))
    print("平均绝对值误差",median_absolute_error(y_test, y_pred_2))
    print("均方误差",mean_squared_error(y_test,y_pred_2))
    print("=============================================")
```

## 分别使用决策树、神经网络、随机森林、支持向量机对股价进行预测

```python
models={"线性回归":LinearRegression(),
        "决策树":DecisionTreeRegressor(max_depth=4),
        "神经网络":MLPRegressor(random_state=0),
        "随机森林":RandomForestRegressor(),
        "支持向量机":SVR(kernel='rbf',C=1000)
       }

for name,model in models.items():
    build_model(name,model)
```

    将上述四种模型与原始模型进行比较，可以发现以时间序列对股价的回归预测都不甚理想，不能获得很好的预测结果。
    纵观上述模型，神经网络在预测过程中表现最佳，可以大致预测出股价变化的趋势，但在具体变化率的预测还有不准。
